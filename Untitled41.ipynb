```python
alls = "Hive Pig HBase Hue Tez Yam Zookeeper Mahout Flume Sqoop Oozie Whirr Storm Apache_Ganglia Apache Falcon WebHDFS NFS Spark Accumulo Knox Tez Spark Cascading Pig MapReduce GraphX MLLib Mahout Drill Shark Impala Hive Accumulo Solr HBase Storm Hue Flume Sqoop Knox Sentry Falcon Oozie Whirr Zookeeper Hive Pig HBase Hue Avro Whirr Flume Yarn Mahout Cloudera_Impala Cloudera_Manager Sqoop Oozie Zookeeper Apache_Sentry"
alls = alls.split()
alls = pd.DataFrame({'a' : alls})
alls = alls.drop_duplicates()
alls['alls']=alls
alls.head()
```

```python
result = pd.merge(alls, hortonWorks, on="a", how="left")
result = pd.merge(result, Cloudera, on="a", how="outer")
result = pd.merge(result, mapr, on="a", how="outer")
print(len(result))
result = result.drop(['a'], axis=1)
```

```python
def movefront(variables,df):
        cols = df.columns.tolist()
        remaining_cols = set(cols) - set(variables) 
        remaining_cols = list(remaining_cols)
        new_cols = variables + remaining_cols
        df = df[new_cols] 
        return df
```
